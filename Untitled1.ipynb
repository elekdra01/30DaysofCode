```
#seperate classes to sexism ,racism and neutral
sexism =  res.loc[res['Class_x'] == 'sexism']
racism = res.loc[res['Class_x'] == 'racism']
none = res.loc[res['Class_x'] == 'none']
```

```
s_tweets = list(sexism.Tweets)
r_tweets = list(racism.Tweets)
n_tweets = list(none.Tweets)
class_list = [s_tweets, r_tweets, n_tweets]
```

```

punctuation = [':',';','!',',','.']
```

```
import re as re
```

```
final_dataset=[]
for i in class_list:
    doc = []
    doc2 = []
    for d in i:
        d = re.sub('<.*?>','',d)
        d = re.sub('\@\w+', '', d)
        d = re.sub('\#\w+','', d)
        d = re.sub('\#','',d)
        d = re.sub('RT','',d)
        d = re.sub('&amp;','',d)
        d = re.sub('[0-9]+','',d)
        d = re.sub('//t.co/\w+','',d)
        d = re.sub('w//','',d)
        d = d.lower()
        doc.append( nltk.word_tokenize( d ) )
    for j in doc:
        for s in j:
            if s not in punctuation:
                doc2.append(s)
    final_dataset.append(doc2)
```

```
#remove stopwords
import nltk
from nltk.corpus import stopwords
stop=stopwords.words('english')
print(stop)
```

```
stop=stopwords.words('english')
notstop=['no','nor','not','never',"isn't" ,"aren't","wasn't" ,"weren't","hasn't" , "haven't" ,"hadn't","doesn't" ,"don't", "didn't","won't" ,
         "wouldn't","shan't","shouldn't" ,"can't", "cannot" ,"couldn't" , "mustn't"]
for ele in notstop:
  if ele in stop:
    stop.remove(ele)
for i in final_dataset:
    for j in i:
        if j in stop:
            i.remove(j)
```

```
#stemmming
from nltk.stem import PorterStemmer
##from nltk.stem import LancasterStemmer not efficient
porter = PorterStemmer()
for i in range( 0, len( final_dataset ) ):
    for j in range( 0, len( final_dataset[ i ] ) ):
        final_dataset[ i ][ j ] = porter.stem( final_dataset[ i ][ j ] )
```

```
sexist = pd.DataFrame({'Sexist tokens': final_dataset[0]})
racist = pd.DataFrame({'Racist tokens': final_dataset[1]})
other = pd.DataFrame({'Other':final_dataset[2]})

racist.to_csv('racist_tokens.csv')
sexist.to_csv('sexist_tokens.csv')
other.to_csv('other_tokens.csv')
```
